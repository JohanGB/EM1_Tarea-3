# Ahora, para una grilla de 0.01 cm se obtiene que las diferencias porcentuales:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the grid
nx, ny = (101, 101) # Change the number of points to match the new grid size
x = np.linspace(0, 10, nx)
y = np.linspace(10, 0, ny)
xv, yv = np.meshgrid(x, y)

# Set the boundary conditions
V = np.zeros((ny, nx))
V[0,:] = 5
V[:,0] = V[:,-1] = V[-1,:] = 0

# Calculate the potential using the finite difference method
for i in range(10000):
    V[1:-1,1:-1] = (V[2:,1:-1] + V[:-2,1:-1] + V[1:-1,2:] + V[1:-1,:-2])/4

# Calculate the potential with a grid size of 0.1
nx2, ny2 = (11, 11)
x2 = np.linspace(0, 10, nx2)
y2 = np.linspace(10, 0, ny2)
xv2, yv2 = np.meshgrid(x2, y2)

V2 = np.zeros((ny2, nx2))
V2[0,:] = 5
V2[:,0] = V2[:,-1] = V2[-1,:] = 0

for i in range(10000):
    V2[1:-1,1:-1] = (V2[2:,1:-1] + V2[:-2,1:-1] + V2[1:-1,2:] + V2[1:-1,:-2])/4

# Calculate the percentage difference between the two potentials
percent_diff = (V2.max() - V.max())/V.max() * 100

# Plot the potential as a function of x and y for both grids
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.pcolormesh(xv, yv, V)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Potencial eléctrico con una grilla de 0.01')

plt.subplot(122)
plt.pcolormesh(xv2, yv2, V2)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Potencial eléctrico con una grilla de 0.1')

plt.suptitle(f'Diferencia porcentual: {percent_diff}%')
plt.show()
```
